# The OWL - Response time analysis
### Section: User Performances
### Network: ASECNAPhB
### Date Range: from 08/04/2020 to 14/04/2020

```python
import pandas as pd
import matplotlib.pyplot as plt             #visualisation
```

```python
df = pd.read_csv("owl_asecna.csv")
df['totalEventsAPV1'] = df.apply(lambda row: row.availabilityEventsAPV1 + row.continuityEventsAPV1, axis=1)
df['totalEventsLPV200'] = df.apply(lambda row: row.availabilityEventsLPV200 + row.continuityEventsLPV200, axis=1)
df  
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df['responseTimeAPV1'], df['totalEventsAPV1'], s=150, alpha=0.5, label="APV1")
plt.scatter(df['responseTimeLPV200'], df['totalEventsLPV200'], s=150, alpha=0.5, label="LPV200")
plt.xlabel('Response time [s]')
plt.ylabel('Events (Availability + Continuity)')
plt.legend(loc=2)
plt.show()
```
